```
def euclidean_algorithm (a,b):
  if a<b :
    a,b = b,a
  while True:
    r = a%b
    if r == 0 :
      return b

    a,b = b,r
```

```
euclidean_algorithm(17,85)
```

```
def euclidean_recursion(a,b):
    if a<b :
      a,b = b,a
    
    r = a%b
    if r == 0:
        return b
    return euclidean_recursion(b,r)
```

```
euclidean_recursion(15,85)
```

```
def selection_sort(array):
  x = array.copy()
  n = len(x)
  for i in range(n):
    min_idx = i
    for j in range(i,n):
      if x[j] < x[min_idx]:
        min_idx = j
      x[i],x[min_idx] = x[min_idx],x[i]
  return x
```

```
random.shuffle(my_array)
my_array
```

```
selection_sort(my_array)
```

```
def factorial(n):
  res = 1
  while n:
    res *= n
    n -= 1
  return res
```

```
factorial(5)
```

```
def factorial_rec(n):
    if n == 1:
        return 1
    return n * factorial_rec(n-1)
```

```
factorial_rec(5)
```
